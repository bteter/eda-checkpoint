```python
#episodes that are have an IMDB rating higher than the average rating (7.15)
high_rating = df[df['imdb_rating'] > 7.15]

#sort table by imdb rating, the highest ratings are on top. 
high_rating = high_rating.sort_values(by='imdb_rating', ascending=False)
print(high_rating.head(10).to_string()) #print top 10 results

# I now have a list of the top 10 episodes of the Simpsons, according to IMDB ratings! 
# Something to note here, every episode in this list is within the first 8 seasons of the show. 
```

```python
# episodes that are have an IMDB rating lower than the average rating

low_rating = df[df['imdb_rating'] < 7.15]
low_rating = low_rating.sort_values(by='imdb_rating', ascending=True)
print(low_rating.head(10).to_string()) #print top 10 results

#List of the lowest rated episodes of the Simpsons.
#Episodes are mostly in later seasons of the show. season 23 and later.
#outliers here are "All Singing, All Dancing" from season 9, and "Gump Roast" from season 13. BOTH episodes are "clip shows". A compilation of previously seen moments of the show. 
```

```python
#average rating of each season, per IMDB rating. Runs average of all episodes in a season

season_avg_ratings = df.groupby('season')['imdb_rating'].mean().reset_index() #group df by season, calculate IMDB average. reset index
season_avg_ratings
#if 7.15 is the average rating, season 1-11 are above average. Every season from 12 on is below average. 
```

```python
#reminder of average viewership
avg_total_views = df["us_viewers_in_millions"].mean()
avg_total_views

#10.38 million
```

```python
# Episodes with higher than average US viewership numbers
# average us viewership numbers for an episode is 10.3 million

high_viewership = df[df['us_viewers_in_millions'] > 10.38] #episodes with viewership higher than 10.3 million.
high_viewership = high_viewership.sort_values(by='us_viewers_in_millions', ascending=False) #list highest to lowest

print(high_viewership.head(10).to_string())

# Top 10 most viewed episodes
# 7 out of 10 of these episodes are in season 1.
# 'Bart gets an F' is the most viewed episode
```

```python
low_viewership = df[df["us_viewers_in_millions"] < 10.38]
low_viewership= low_viewership.sort_values(by='us_viewers_in_millions', ascending=True)

print(low_viewership.head(10).to_string())

# top 10 lowest viewed episodes
# 'Hostile Kirk Place' is the episode with the lowest amount of viewers
# All episodes are from the last 2 years
```

```python
#Seasons with above average viewership numbers 

season_avg_viewers = df.groupby('season')['us_viewers_in_millions'].mean().reset_index() #group df by season, calculate average us viewer. reset index
above_avg_seasons = season_avg_viewers[season_avg_viewers["us_viewers_in_millions"] > 10.38] #only show seasons with viewers higher than the average 

above_avg_seasons

#first 10 seasons are consistently above average
#this streak in above average viewership skips season 11, but continues with seasons 12-15 and stops there.
```

```python
#seasons with below average viewers

below_avg_seasons = season_avg_viewers[season_avg_viewers["us_viewers_in_millions"] < 10.38]
below_avg_seasons

#decline in viewership begins with season 16.
```

```python
#compare viewers and IMDB rating

views_ratings_combo = pd.merge(season_avg_viewers, season_avg_ratings, on='season') #merge my table showing average episode rating with table showing average IMDB rating. Merge on shared 'season' table
print(views_ratings_combo.to_string())
```

## Visualize

Create any visualizations for your EDA here. Make note in the form of code comments of what your thought process is for your visualizations.

```python
#Average episode rating per season visualization
import matplotlib.pyplot as plt # I was getting an error "module 'matplotlib' has no attribute 'figure'"", according to a google search i needed to import matplotlib this way.

season_avg_ratings = df.groupby('season')['imdb_rating'].mean().reset_index() #group df by season, calculate IMDB average. reset index
# print(season_avg_ratings_plot)

# # color codes (blue and yellow) from the Simpsons' color palette (from https://www.color-hex.com/color-palette/12626)
simpsons_yellow = "#F8DB27" 
simpsons_blue = '#2F64D6' 

plt.figure(figsize=(10, 6)) #set up line graph, set graph's size
plt.plot(season_avg_ratings['season'], season_avg_ratings['imdb_rating'], marker='o', color=simpsons_yellow) #x-axis = season, y-axis = imdb rating

#labels
plt.title('Average IMDb Ratings per Season of The Simpsons Season 1 - 34', color=simpsons_blue)
plt.xlabel('Season', color=simpsons_blue)
plt.ylabel('Average IMDb Rating', color=simpsons_blue)

plt.grid(True) #add grid lines
plt.show()

#ratings start to decline at season 7, and dip below average at season 15
```
